```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Read the dataset to the python environment.

```python
data=pd.read_csv('cars_data.csv')
```

# 3. Plot a bar graph of male vs female buyers participated in the sales.


```python
sns.barplot(x='Buyer Gender',y='Sale Price',data=data)
plt.show()
```

# 4. Find the top 5 cars based on their sales price.

```python
data.Make.value_counts().nlargest(5).plot(kind='bar',figsize=(10,5))
plt.title('Growth Index')
plt.xlabel('Make')
plt.ylabel('Sale Price')
plt.show()
```

# 5. Find the least 5 cars based on their Resell price.

```python
data.Make.value_counts().nsmallest(5).plot(kind='bar',figsize=(10,5))
plt.title('Growth Index')
plt.xlabel('Make')
plt.ylabel('Resell Price')
plt.show()
```
